```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

```python
auto = pd.read_csv('Social_Network_Ads.csv')
```

```python
auto.head()
```

```python
sns.distplot(auto['EstimatedSalary'])
plt.show()
```

```python
sns.distplot(auto['EstimatedSalary'], kde=False, rug=True)
plt.show()
```

```python
sns.pairplot(auto[['EstimatedSalary', 'Gender', 'Age']])
```

```python
sns.stripplot(auto['Gender'], auto['Age'], jitter = True)
```

```python
sns.swarmplot(auto['Gender'], auto['Age'])
```

```python
sns.boxplot(auto['EstimatedSalary'], auto['Gender'], hue=auto['Age'])
```

```python
sns.barplot(auto['EstimatedSalary'], auto['Gender'], hue= auto['Age'])
```

```python
sns.countplot(auto['Gender'], hue=auto['Age'])
```

```python
sns.pointplot(auto['EstimatedSalary'], auto['Gender'], hue=auto['Age'])
```

```python
sns.lmplot(x='Age', y ='EstimatedSalary', data= auto, hue='Gender')
```
